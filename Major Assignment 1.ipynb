### 4. Write a python program to find the mean, median and mode of weights of users.


```python

def mean(numbers):
    return sum(numbers)/len(numbers)

def median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 0:
        return (sorted_numbers[n//2-1] + sorted_numbers[n//2])/2
    else:
        return sorted_numbers[n//2]
def mode(numbers):
    d = {}
    for item in numbers:
        if item in d:
            d[item] += 1
        else:
            d[item] = 1
    for key in d.keys():
        if d[key] == max(d.values()):
            return key 
mean_weight = mean(weights)
median_weight = median(weights)   |
mode_weight = mode(weights)
mean_weight,median_weight,mode_weight
```
